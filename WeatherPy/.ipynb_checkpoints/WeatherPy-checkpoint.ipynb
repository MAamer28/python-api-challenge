## Starter Code to Generate Random Geographic Coordinates and a List of Cities

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Impor the OpenWeatherMap API key
from api_keys import weather_api_key

# Import citipy to determine the cities based on latitude and longitude
from citipy import citipy
```

### Generate the Cities List by Using the `citipy` Library

```python
# Empty list for holding the latitude and longitude combinations
lat_lngs = []

# Empty list for holding the cities names
cities = []

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(f"Number of cities in the list: {len(cities)}")
```

---

## Requirement 1: Create Plots to Showcase the Relationship Between Weather Variables and Latitude

### Use the OpenWeatherMap API to retrieve weather data from the cities list generated in the started code

```python
# Set the API base URL
url = "http://api.openweathermap.org/data/2.5/weather?appid=" + weather_api_key + "&q="

# Define an empty list to fetch the weather data for each city
city_data = []

# Print to logger
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop through all the cities in our list to fetch weather data
for i, city in enumerate(cities):
        
    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 0

    # Create endpoint URL with each city
    city_url = url + city + "&units=metric"
    
    # Log the url, record, and set numbers
    print("Processing Record %s of Set %s | %s" % (record_count, set_count, city))

    # Add 1 to the record count
    record_count += 1

    # Run an API request for each of the cities
    try:
        # Parse the JSON and retrieve data
        response_json = requests.get(city_url).json()

        # Parse out latitude, longitude, max temp, humidity, cloudiness, wind speed, country, and date
        city_lat = response_json["coord"]["lat"]
        city_lng = response_json["coord"]["lon"]
        city_max_temp = response_json["main"]["temp_max"]
        city_humidity = response_json["main"]["humidity"]
        city_clouds = response_json["clouds"]["all"]
        city_wind = response_json["wind"]["speed"]
        city_country = response_json["sys"]["country"]
        city_date = response_json["dt"]

        # Append the City information into city_data list
        city_data.append({"City": city, 
                          "Lat": city_lat, 
                          "Lng": city_lng, 
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})

    # If an error is experienced, skip the city
    except:
        print("City not found. Skipping...")
        pass
              
# Indicate that Data Loading is complete 
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Convert the cities weather data into a Pandas DataFrame
city_data_df = pd.DataFrame(city_data)


# Show Record Count
city_data_df.count()
```

## Requirement 2: Compute Linear Regression for Each Relationship


```python
# Define a function to create Linear Regression plots
def plot_linear_regression(x_values, y_values, title, text_coordinates):    
    (slope, intercept, r_value, p_value, sem) = linregress (x_values, y_values)
    regress_val = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + " x + " + str(round(intercept,2))
    
    plt.scatter(x_values, y_values)
    plt.plot(x_values, regress_val, "r-")
    plt.annotate(line_eq, text_coordinates, fontsize = 15)
    plt.xlabel("Latitude")
    print(f"The r-squared is: {r_value}")
    plt.show()
```

```python
# Create a DataFrame with the Northern Hemisphere data (Latitude >= 0)
northern_hemi_df = city_data_df.loc[(city_data_df["Lat"] >= 0)]
    
# Display sample data
northern_hemi_df.head()
```

```python
# Create a DataFrame with the Southern Hemisphere data (Latitude < 0)
southern_hemi_df = city_data_df.loc[(city_data_df["Lat"] <0)]

# Display sample data
southern_hemi_df.head()
```

###  Temperature vs. Latitude Linear Regression Plot

```python
# Linear regression on Northern Hemisphere

x_values = northern_hemi_df["Lat"]
y_values = northern_hemi_df["Max Temp"]
plot_linear_regression(x_values, y_values, "Max Temp", (40, 40))
```

```python
# Linear regression on Southern Hemisphere

x_values = southern_hemi_df["Lat"]
y_values = southern_hemi_df["Max Temp"]
plot_linear_regression(x_values, y_values, "Max Temp", (-50, 30))
```

**Discussion about the linear relationship:** 

The data suggests there is a clear and strong relationship between latitude and temperature. Cities closer to the equator tend to have higher maximum temperatures than cities further to the north or south, as both hemispheric analyses have shown. 

Temperatures in cities in the northern hemisphere show an r-squared value of -0.71, indicating a strong negative linear relationship between city latitude and temperature. On the other hand, temperatures in southern hemisphere cities show a less strong but still pronounced relationship to latitude, with a positive linear r-squared value of 0.48.

The negative and positive nature of the r-squared values reflects the latitude-longitude system whereby the northern hemisphere takes the equator as 0 and latitudes increase further north from the equator. Southern latitudes on the other hand take the equator as 0 but move further into negative values as they go further south. As such, the r-squared values reflect this system.

### Humidity vs. Latitude Linear Regression Plot

```python
# Northern Hemisphere

x_values = northern_hemi_df["Lat"]
y_values = northern_hemi_df["Humidity"]
plot_linear_regression(x_values, y_values, "Humidity", (40, 10))
```

```python
# Southern Hemisphere

x_values = southern_hemi_df["Lat"]
y_values = southern_hemi_df["Humidity"]
plot_linear_regression(x_values, y_values, "Humidity", (-55, 20))
```

**Discussion about the linear relationship:** 

According to the data, there does not appear to be any significant relationship between latitude and humidity. This matches general understanding of humidity as a function of more than just latitude and temprature, with complex interactions of pressure systems and oceanic currents determining the humidity of a given city.

As such, cities closer to the equator may have higher humidity indices, indicating a tropical climate, or they could be drier, indicating arid conditions. Both tropical and desert climates exist in areas closer to the equator, with the Namib, Sahara, and Arabian deserts coexisting in the latitudes between the two tropics along with the Amazon, Congo, and Borneo rainforests.

Hence, there is little to be said about a univariant relationship between latitude and humidity.
